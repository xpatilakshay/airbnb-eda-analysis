## Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("airbnb.csv")
```

## Checking Missing Values

```python
data.isnull().sum()
```

## Handling Missing Values

```python
# Converting last reviews data type into date

data['last review'] = pd.to_datetime(data['last review'],errors='coerce')
```

```python
# filling Nan Data

data.fillna({'reviews per month':0,'last review':data['last review'].min()},inplace=True)
```

```python
# Dropping null values

data.dropna(subset=['NAME','host name'],inplace=True)
```

```python
data.isnull().sum()
```

```python
# Dropping Unneccessary Columns
data = data.drop(columns=['license','house_rules'],errors='coerce')
```

```python
# Removing $ unit from price and service fee

data['price'] = data['price'].replace('[\$,]','',regex=True).astype(float)
data['service fee'] = data['service fee'].replace('[\$,]','',regex=True).astype(float) 
```

```python
data
```

## Removing Duplicates

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

## Visualization

### What is the distribution of listing prices?

```python
plt.hist(data['price'],bins=50,color='red',edgecolor='Black',width=25)
plt.title('Distribution of Listing Prices')
plt.xlabel('Price')
plt.ylabel('Number of Listings')
plt.show()
plt.show()
```

**Conclusion:**
The histogram displays the distribution of listing prices with higher granularity due to 50 bins. Most listings appear to be concentrated in the lower price ranges, indicating that budget-friendly accommodations are more common. As the price increases, the frequency of listings gradually decreases, suggesting fewer high-priced properties. This pattern highlights a right-skewed distribution, which is typical in pricing data where luxury listings are less frequent.

## How are different rooms types distributed?

```python
total_rooms = data['room type'].value_counts()
plt.bar(total_rooms.index,total_rooms,color='red')
plt.title('Count of Listings by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.show()
plt.show()
```

## How are listings distributed across different neighborhoods?

```python
# total_neighbour = data['neighbourhood group'].value_counts()
# plt.barh(total_neighbour.index,total_neighbour.values,color='orange')
# plt.title('Distribution of Listings by Neighbourhood Group')
# plt.xlabel('Number of Listings')
# plt.ylabel('Neighbourhood Group')
# plt.show()


# in ascending

total_neighbour = data['neighbourhood group'].value_counts().sort_values(ascending=True)
plt.barh(total_neighbour.index,total_neighbour.values,color='orange')
plt.title('Distribution of Listings by Neighbourhood Group')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood Group')
plt.show()
```

## what is the relationship between price and room type?

```python
data1 = data.groupby('room type')['price'].mean()
plt.bar(data1.index,data1.values,color='orange')
plt.bar(data1.index, data1.values, color='orange')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.xticks(rotation=15) 
plt.show()
```

## how has the number of reviews change over with time

```python
data['last review'] = pd.to_datetime(data['last review'])
reviews_over_time = data.groupby(data['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
